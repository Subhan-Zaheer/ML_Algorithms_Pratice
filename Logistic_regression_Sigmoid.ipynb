```python
def logistic_reg(X, y):
    predictions = np.zeros(100)
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        idx = np.random.randint(X.shape[0])
        y_pred = sigmoid(np.dot(X[idx], weights))
        y_pred= decision(y_pred)
        predictions[idx] = y_pred
        weights = weights + lr*(y[idx]- y_pred)*X[idx]
    
    return weights[0], weights[1:], predictions
        
```

```python
def decision(x):
    return 1 if x > 0.5 else 0
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-1*z))
```

```python
intercept_ , coef_, predictions = logistic_reg(X, y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/ coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + b
y_input_1 = m1*x_input + b1
```
