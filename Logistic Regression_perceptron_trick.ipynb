```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2,
                          n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:,-1], c=y, cmap='winter', s=100)
```

```python
def perceptron_trick(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        idx = np.random.randint(0, 100)
        
        y_pred = decision(np.dot(X[idx],weights))
        weights = weights + lr*(y[idx] - y_pred)*X[idx]
        
    
    return weights[0], weights[1:]
        
        
        
```

```python
def another_perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.01
    
    for i in range(1000):
        idx = np.random.randint(0, 100)
#         print(X[idx].shape, weights.shape)
        temp = np.dot(X[idx], weights)
        print(f"Shape of temp is : {temp.shape}")
#         y_pred = decision(np.dot(X[idx], weights))
#         x = np.sum(temp)
#         print(x)
        y_pred = decision(temp)
        if y[idx] == 0 and y_pred == 1:
            weights = weights - (lr*X[idx])
            
        elif y[idx] == 1 and y_pred == 0:
            weights = weights - (lr*X[idx])
        
    return weights[0], weights[1:]
        
        
```

```python
def decision(x):
    return 1 if x >= 0 else 0
```

```python
intercept_, coef_ = perceptron_trick(X, y)
```
