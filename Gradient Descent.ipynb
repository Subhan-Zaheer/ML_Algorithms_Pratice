```python
from sklearn.datasets import make_regression
import numpy as np
from sklearn.metrics import r2_score
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1 ,n_targets=1, noise=50, random_state=12)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
#Lets apply formula of Linear Regression

from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(X, y)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, LR.predict(X), color='red')
```

```python
pred = LR.predict(X)
```

```python
r2_score(X, pred)
```

```python
# Now applying Gradient Descent assuming that slope is constant = 1.77
# and let's assume the starting value of b = 0

y_pred = ((1.77 * X)+ 0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X, LR.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='#00a65a', label='b = 0')
plt.legend()
```

```python
m = 1.77
b = 0

loss_slope = -2 * np.sum(y-m*X.ravel() - b)
loss_slope
```

```python
learning_rate = 0.1

step_size = learning_rate * loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((1.77 * X)+b).reshape(4)

plt.scatter(X, y)
plt.plot(X, LR.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color = '#00a65a', label='b = 0')
plt.plot(X, y_pred1, color='#A3E4D7', label=f'b = {b}')
plt.legend()
plt.show()
```

```python
b = 20
m = 1.77
learning_rate = 0.01
epochs = 20

for i in range(epochs):
    loss_slope = -2 *np.sum(y - m * X.ravel() - b)
    b = b - (loss_slope * learning_rate)
    
    y_pred = m * X + b
    
    plt.plot(X, y_pred)

plt.scatter(X, y)
    
    
```
